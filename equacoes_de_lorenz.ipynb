```python
# importando as bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# definindo o método computacional
def metodo(f, condicoes_iniciais, dt, n):
  x, y, z = condicoes_iniciais
  trajetoria = np.zeros((n, 3))

  for i in range(n):
    trajetoria[i] = [x, y, z]
    dx, dy, dz = f(x, y, z)
    x = x + dx * dt
    y = y + dy * dt
    z = z + dz * dt

  return trajetoria, n
```

```python
# definindo a equação de lorenz
def lorenz(x, y, z):
  sigma = 10
  rho = 28
  beta = 8/3

  dx = sigma * (y - x)
  dy = x * (rho - z) - y
  dz = x * y - beta * z

  return dx, dy, dz
```

```python
condicoes_iniciais = [0.1, 0.0, 0.0]

trajetoria, n = metodo(lorenz, condicoes_iniciais, 0.01, 10000)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for i in range(n - 1):
    ax.plot(trajetoria[i:i+2, 0], trajetoria[i:i+2, 1], trajetoria[i:i+2, 2], c=plt.cm.viridis(i/n))
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Trajetória das Equações de Lorenz')
plt.show()
```
